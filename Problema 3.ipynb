```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la lognormal calculados a partir de la media y desviación estándar dadas
m = 10  # media en minutos
s = 2   # desviación estándar en minutos

# Cálculo de los parámetros de la distribución normal subyacente
mu_ln = np.log(m**2 / np.sqrt(s**2 + m**2))
sigma_ln = np.sqrt(np.log(1 + (s**2 / m**2)))

# Generar tiempos de espera para 50 clientes usando la distribución lognormal
tiempos_espera = np.random.lognormal(mean=mu_ln, sigma=sigma_ln, size=50)

# Mostrar los tiempos de espera
print("Tiempos de espera de 50 clientes (en minutos):")
for i, tiempo in enumerate(tiempos_espera, start=1):
    print(f"Cliente {i}: {tiempo:.2f} minutos")

# Graficar los tiempos de espera
plt.figure(figsize=(10, 6))
plt.hist(tiempos_espera, bins=15, color='lightcoral', edgecolor='black')
plt.xlabel("Tiempo de espera (minutos)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Tiempos de Espera de los Clientes en el Servicio Técnico")
plt.show()
```
