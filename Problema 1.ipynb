```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución binomial
n_pedidos = 20  # Número de pedidos diarios
prob_exito = 0.7  # Probabilidad de cumplir un pedido
dias_semana = 7  # Número de días en una semana

# Simulación del número de pedidos satisfechos por día durante una semana
pedidos_satisfechos_semanal = np.random.binomial(n=n_pedidos, p=prob_exito, size=dias_semana)

# Mostrar los resultados por día
print("Pedidos satisfechos cada día durante una semana:")
for dia, pedidos in enumerate(pedidos_satisfechos_semanal, start=1):
    print(f"Día {dia}: {pedidos} pedidos satisfechos")

# Gráfica de barras
plt.figure(figsize=(10, 6))
plt.bar(range(1, dias_semana + 1), pedidos_satisfechos_semanal, color='skyblue')
plt.xlabel("Día")
plt.ylabel("Pedidos Satisfechos")
plt.title("Número de Pedidos Satisfechos por Día durante una Semana")
plt.xticks(range(1, dias_semana + 1))
plt.show()
```
