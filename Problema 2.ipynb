```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución exponencial
tasa_llegadas_por_hora = 4  # tasa de 4 llegadas por hora
tiempo_operacion_horas = 12  # tiempo de operación en horas

# Generar tiempos entre llegadas
tiempos_entre_llegadas = []
tiempo_acumulado = 0

# Generar tiempos hasta completar 12 horas de operación
while tiempo_acumulado < tiempo_operacion_horas:
    tiempo_entre_llegadas = np.random.exponential(scale=1/tasa_llegadas_por_hora)
    tiempos_entre_llegadas.append(tiempo_entre_llegadas)
    tiempo_acumulado += tiempo_entre_llegadas

# Mostrar los tiempos entre llegadas
print("Tiempos entre llegadas de vehículos (en horas):")
for i, tiempo in enumerate(tiempos_entre_llegadas, start=1):
    print(f"Llegada {i}: {tiempo:.2f} horas")

# Gráfica de los tiempos entre llegadas
plt.figure(figsize=(10, 6))
plt.hist(tiempos_entre_llegadas, bins=20, color='lightgreen', edgecolor='black')
plt.xlabel("Tiempo entre llegadas (horas)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Tiempos entre Llegadas de Vehículos")
plt.show()
```
